Understand how Delta Lake gives ACID guarantees on top of plain object storage (S3/DBFS).

Delta does NOT modify files. It logs what happened.

That log = `_delta_log`.

## Step 1: Create a Delta Table

```python
%sql

CREATE SCHEMA IF NOT EXISTS lab_2026;
USE lab_2026;

CREATE OR REPLACE TABLE events (
  id INT,
  data STRING
) USING DELTA;
```

```python
%sql

DESC FORMATTED events;
```

## Step 2: Insert Data

```python
%sql

INSERT INTO events VALUES (1, 'click');
INSERT INTO events VALUES (2, 'view');
INSERT INTO events VALUES (3, 'purchase');
```

## Step 3: Inspect Delta Log

```python
%sql

DESCRIBE HISTORY events;
```
