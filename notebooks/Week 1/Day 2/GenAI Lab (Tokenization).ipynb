## Step 2Ô∏è‚É£ ‚Äî Load a tokenizer

````python
enc = tiktoken.get_encoding("cl100k_base")
````

## Step 3Ô∏è‚É£ ‚Äî Tokenize simple text vs number

````python
text_word = "Data"
text_number = "794625452451"

print("Word tokens:", enc.encode(text_word))
print("Number tokens:", enc.encode(text_number))

print("Word token count:", len(enc.encode(text_word)))
print("Number token count:", len(enc.encode(text_number)))
````

**Output:**
```
Word tokens: [1061]
Number tokens: [25926, 15894, 21098, 20360]
Word token count: 1
Number token count: 4
```

## üß† WHY THIS HAPPENS (This is the core insight)
**LLM tokenizers are:**
- Optimized for natural language
- Trained on common word patterns
- BAD at long, random-looking numbers

**So:**
- `Data` ‚Üí seen millions of times ‚Üí single token
- `794625452451` ‚Üí rare pattern ‚Üí split into chunks
